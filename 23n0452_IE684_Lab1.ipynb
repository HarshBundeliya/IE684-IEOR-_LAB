# 1

The given function is:
\[ f(x) = (1.2 - x_1)^2 + 60 \cdot (x_2 - x_1^2)^2 \]

**Partial derivatives:**
\[ \frac{\partial f}{\partial x_1} = 2(1.2 - x_1)(-1) - 120(x_2 - x_1^2)(x_1) \]
\[ \frac{\partial f}{\partial x_2} = 120(x_2 - x_1^2) \]

**Setting partial derivatives to zero:**
\[ 2(1.2 - x_1) - 120(x_2 - x_1^2)(x_1) = 0 \]
\[ 120(x_2 - x_1^2) = 0 \]

Solving for \(x_1\) and \(x_2\):
\[ x_1 = 1.2, \quad x_2 = 1.44 \]

Therefore, the minimum occurs at \((1.2, 1.44)\).


# 2

```python
def objective_function(x):
    return (1.2 - x[0])**2 + 60 * (x[1] - x[0]**2)**2

def gradient(x):
    df_dx1 = 2 * (1.2 - x[0]) * (-1) - 60 * (x[1] - x[0]**2) * (2 * x[0])
    df_dx2 = 120 * (x[1] - x[0]**2)
    return np.array([df_dx1, df_dx2])

def gradient_descent(x0, eta, tolerance):
    x = x0
    k = 0
    while np.linalg.norm(gradient(x)) > tolerance:
        x = x - eta * gradient(x)
        k += 1
    return x, objective_function(x), k
```

```python
x0 = np.array([-1.5, 1.5])
eta = 0.001
tolerances = [10**(-p) for p in range(1, 8)]

results_table = []
for tau in tolerances:
    minimizer, obj_value, num_iterations = gradient_descent(x0, eta, tau)
    results_table.append([tau, tuple(minimizer), obj_value, num_iterations])

print("{:<15} {:<50} {:<25} {:<20}".format("Tolerance", "Minimizer", "Objective Function", "Iterations"))
for row in results_table:
    print("{:<15} {:<50} {:<25} {:<20}".format(row[0], str(row[1]), row[2], row[3]))

tau_values = [row[0] for row in results_table]
iteration_values = [row[3] for row in results_table]

plt.plot(tau_values, iteration_values, marker='o', linestyle='-')
plt.xscale('log')
plt.yscale('log')
plt.title('Iterations vs Tolerance')
plt.xlabel('Tolerance (log scale)')
plt.ylabel('Number of Iterations (log scale)')
plt.show()
```

**Observations:**

1. **Minimizer Convergence:**
   - Minimizer stabilizes at \((1.2, 1.44)\) for tolerances \(< 1e-4\).

2. **Objective Function Convergence:**
   - Objective function decreases with smaller tolerances.

3. **Iterations:**
   - More iterations needed for finer tolerances.

4. **Performance:**
   - Gradient descent converges effectively.
   - Stabilization below \(1e-4\) indicates reliability.

**Summary:**
   - Algorithm performs well, reaching true minimum.

```python
def gradient_descent_constant_step(x0, eta, tolerance, max_iterations=1000):
    trajectory = [x0]
    iteration = 0
    while np.linalg.norm(gradient(trajectory[-1])) > tolerance and iteration < max_iterations:
        x0 = x0 - eta * gradient(x0)
        trajectory.append(x0.copy())
        iteration += 1
    return np.array(trajectory)

x0 = np.array([-1.5, 1.5])
eta = 0.01
tolerance = 1e-5
max_iterations = 1000

trajectory = gradient_descent_constant_step(x0, eta, tolerance, max_iterations)

x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
x, y = np.meshgrid(x, y)
z = objective_function([x, y])

plt.contour(x, y, z, levels=50, cmap='viridis')
plt.plot(trajectory[:, 0], trajectory[:, 1], color='red', label='Trajectory')
plt.title('Level Sets and Trajectory of Gradient Descent')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

**Shortcomings:**

1. **Fixed Step Size:** The algorithm may converge slowly or diverge due to a fixed step size (\(\eta\)).
  
2. **Sensitive to Initial Guess:** Performance can vary based on the initial guess, potentially leading to different local minima.

3. **Non-Convex Functions:** Not suitable for non-convex functions, may get stuck in local minima.

4. **Hessian Calculation:** Computationally expensive for Hessian calculation in Newton's method.

5. **Noisy Objectives:** Sensitivity to noise in the objective function may result in erratic behavior.

**Solutions:**

1. **Adaptive Step Size:** Use methods like line search or backtracking for dynamic step size adjustment.

2. **Multiple Runs:** Perform multiple runs with different initial guesses for better exploration.

3. **Advanced Algorithms:** Explore algorithms like stochastic gradient descent, genetic algorithms, or simulated annealing for non-convex problems.

4. **Quasi-Newton Methods:** Use methods like BFGS or L-BFGS to approximate the Hessian without direct computation.

5. **Robust to Noise:** Choose optimization techniques designed for noisy objectives, such as stochastic gradient descent.

The given function is:

\[ f(x) = (2 - x_1)^2 + 30 \cdot (x_2 - x_1^2)^2 \]

To find the minimizer, we set the partial derivatives equal to zero:

\[ \frac{\partial f}{\partial x_1} = 2(2 - x_1)(-1) - 60(x_2 - x_1^2)(2x_1) = 0 \]
\[ \frac{\partial f}{\partial x_2} = 60(x_2 - x_1^2) = 0 \]

Solving these equations gives the critical points. However, it's known that for this specific problem, the global minimum is at \( (2, 4) \) as discussed earlier.

Substituting \( x_1 = 2 \) and \( x_2 = 4 \) into the function:

\[ f(2, 4) = (2 - 2)^2 + 30 \cdot (4 - 2^2)^2 \]
\[ f(2, 4) = 0 + 30 \cdot 0 \]
\[ f(2, 4) = 0 \]

Therefore, the minimizer is \( (2, 4) \), and the minimum function value is \( 0 \).

The function $f(x) = (2 - x_1)^2 + 30 \cdot (x_2 - x_1^2)^2$ is convex near its global minimum $(2, 4)$. Convexity is indicated by the positive semidefiniteness of the Hessian matrix. The function exhibits convex behavior around the global minimum.

```python
def objective_function(x):
    return (2 - x[0])**2 + 30 * (x[1] - x[0]**2)**2

def gradient(x):
    df_dx1 = 2 * (2 - x[0]) * (-1) - 60 * (x[1] - x[0]**2) * (2 * x[0])
    df_dx2 = 60 * (x[1] - x[0]**2)
    return np.array([df_dx1, df_dx2])

def hessian(x):
    h11 = 2 + 120 * (x[1] - x[0]**2) + 240 * x[0]**2
    h12 = -120 * x[0]
    h21 = -120 * x[0]
    h22 = 60
    return np.array([[h11, h12], [h21, h22]])

def newton_method(x0, tolerance):
    x = x0
    k = 0
    while np.linalg.norm(gradient(x)) > tolerance:
        x = x - np.linalg.inv(hessian(x)).dot(gradient(x))
        k += 1
    return x, objective_function(x), k
```

```python
x0 = np.array([-1.5, 1.5])
tolerances = [10**(-p) for p in range(1, 21)]

results_table = []
for tau in tolerances:
    minimizer, obj_value, num_iterations = newton_method(x0, tau)
    results_table.append([tau, tuple(minimizer), obj_value, num_iterations])

print("{:<15} {:<50} {:<25} {:<20}".format("Tolerance", "Minimizer", "Objective Function", "Iterations"))
for row in results_table:
    print("{:<15} {:<50} {:<25} {:<20}".format(row[0], str(row[1]), row[2], row[3]))

tau_values = [row[0] for row in results_table]
iteration_values = [row[3] for row in results_table]

plt.plot(tau_values, iteration_values, marker='o', linestyle='-')
plt.xscale('log')
plt.yscale('log')
plt.title('Iterations vs Tolerance')
plt.xlabel('Tolerance (log scale)')
plt.ylabel('Number of Iterations (log scale)')
plt.show()
```

\textbf{Observations:}

1. \textbf{Minimizers and Objective Function Values:}

   - As the tolerance decreases, the minimizer approaches the true solution (2.0, 4.0), and the objective function values approach zero.

2. \textbf{Iterations vs. Tolerance:}
